```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.imdb.com/india/top-rated-indian-movies/"
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
scraped_indmovies = soup.find_all('td', class_='titleColumn')
scraped_indmovies[0:100]
```

```python
movies = []
for movie in scraped_indmovies:
    movie = movie.get_text().replace('\n', "")
    movie = movie.strip(" ")
    movies.append(movie)
movies[0:100]
```

```python
scraped_ratings = soup.find_all('td', class_='ratingColumn imdbRating')
scraped_ratings[0:100]
```

```python
ratings1 = []
for rating in scraped_ratings:
    rating = rating.get_text().replace('\n', '')
    ratings1.append(rating)
ratings1[0:100]
```

```python
data = pd.DataFrame()
data['Movie Names'] = movies
data['Ratings'] = ratings1
data.head()
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.puredestinations.co.uk/top-10-famous-monuments-to-visit-in-india/"
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
data=soup.find_all('div',class_="blog--single__content column--3-4 u-spacing-third")
temp=[]
for i in data:
    for j in i.find_all('p'):
        t= j.text.replace('\n','')
        temp.append(t)
```

```python
name=[]
desc=[]
for i in range(1,30,3):
    name.append(temp[i])
for i in range(2,30,3):
    desc.append(temp[i])
```

```python
name
```

```python
desc
```

```python
temp=[]
for i in soup.find_all('img'):
    temp.append(i['src'])
```

```python
img_url=[]
for i in range(6,26,2):
    img_url.append(temp[i])
```

```python
img_url
```

```python
df = pd.DataFrame()
df['Monument']=name
df['description']=desc
df['Image_url']=img_url
df
```
